```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


# def l_list_to_number(node):
#     curr_node = node
#     char_list = []
#     while curr_node:
#         char_list.append(str(curr_node.val))
#         curr_node = curr_node.next
#     return int(''.join(char_list))


class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        
        ptr_1 = l1
        ptr_2 = l2 
        
        result = ListNode()
        head = result
        
        carry = 0
        
        while (ptr_1 is not None) or (ptr_2 is not None): 

            if ptr_1 is not None: 
                num_1 = ptr_1.val
                ptr_1 = ptr_1.next
            else: 
                num_1 = 0
                
            if ptr_2 is not None:
                num_2 = ptr_2.val
                ptr_2 = ptr_2.next
            else: 
                num_2 = 0
                
            num_3 = num_1 + num_2 + carry
            
            result.next = ListNode(num_3%10)

            if num_3 > 9 : 
                carry = 1
            else: 
                carry = 0
                
            result = result.next

        # finally
        if carry == 1: 
            result.next = ListNode(carry)
            
        return head.next
```
